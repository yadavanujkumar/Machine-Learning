```python
import pandas as pd
```

```python
 dataframe=pd.read_csv("shop.csv")
```

```python
dataframe.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
dataframe['Genre']=lb.fit_transform(dataframe['Genre'])
```

```python
 dataframe.head()
```

```python
import matplotlib.pyplot as plt
```

```python
import scipy.cluster.hierarchy as shc
```

```python
plt.figure(figsize=(25,15))
dendogram1=shc.dendrogram(shc.linkage(dataframe,method='ward'))
```

```python
data=dataframe.iloc[:,2:5]
```

```python
data.head()
```

```python
plt.figure(figsize=(25,15))
dendogram2=shc.dendrogram(shc.linkage(data,method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=5)
```

```python
model.fit_predict(data)
```

```python
plt.figure(figsize=(25,15))
plt.scatter(data['Annual Income (k$)'],data['Spending Score (1-100)'],c=model.labels_)
```
