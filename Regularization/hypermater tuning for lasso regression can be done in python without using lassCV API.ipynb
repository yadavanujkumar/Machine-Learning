hypermater tuning for lasso regression can be done in python without using lassCV API


```python
import pandas as pd
```

```python
df = pd.read_csv("Boston_Housing.csv")
```

```python
df.head(3)
```

```python
x=df.iloc[:,:-1]
```

```python
x.shape
```

```python
y = df.iloc[:,-1]
```

```python
y.shape
```

```python
from sklearn.linear_model import Lasso
model = Lasso()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.25, random_state= 1)
model.fit(xtrain,ytrain)
```

```python
from sklearn.model_selection import RepeatedKFold
cv=RepeatedKFold(n_splits=10,n_repeats=3, random_state=1)
```

```python
from sklearn.metrics import r2_score
ypred=model.predict(xtest)
r2_score(ytest,ypred)
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x_sc = sc.fit_transform(x)
xtrain,xtest,ytrain,ytest = train_test_split(x_sc,y, test_size=0.25 , random_state=1)
model1=Lasso()
params={'alpha':[0.00001,0.0001,0.001,0.01]}

from sklearn.model_selection import GridSearchCV
search = GridSearchCV(model1,params,cv=cv)
result=search.fit(x_sc,y)
result.best_params_
```

```python
model2 = Lasso(alpha=0.1)
model2.fit(xtrain,ytrain)
```

```python
ypred2=model2.predict(xtest)
r2_score(ytest, ypred2)
```
