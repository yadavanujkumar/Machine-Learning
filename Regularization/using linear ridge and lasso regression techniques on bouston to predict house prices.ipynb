```python
import pandas as pd

from sklearn.datasets import fetch_california_housing
boston = fetch_california_housing()
boston_df=pd.DataFrame(boston.data , columns= boston.feature_names)
boston_df['Price_House'] = boston.target
boston_df
```

```python
x=boston_df.iloc[:,:-1]
```

```python
x.shape
```

```python
y = boston_df.iloc[:,-1]
```

```python
y.shape
```

```python
#scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_sc = sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x_sc, y ,test_size=0.3 , random_state=1)
```

```python
#applying linear regression
from sklearn.metrics import r2_score,mean_absolute_error
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(xtrain,ytrain)
```

```python
predict = lr.predict(xtest)
print(r2_score(ytest,predict))
```

```python
print(mean_absolute_error(ytest,predict))
```

```python
#applying linear regression
from sklearn.linear_model import Ridge
rr = Ridge()
rr.fit(xtrain,ytrain)
```

```python
predict1 = rr.predict(xtest)
print(r2_score(ytest,predict1))
```

```python
print(mean_absolute_error(ytest,predict1))
```

```python
#now applying repeated k fold
from sklearn.model_selection import RepeatedKFold
cv=RepeatedKFold(n_splits=10,n_repeats=3, random_state=1)
from sklearn.metrics import r2_score
ypred=rr.predict(xtest)
r2_score(ytest,ypred)

from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x_sc = sc.fit_transform(x)
xtrain,xtest,ytrain,ytest = train_test_split(x_sc,y, test_size=0.25 , random_state=1)
model1=Ridge()
params={'alpha':[0.00001,0.0001,0.001,0.01,0.1,1,5,10]}

from sklearn.model_selection import GridSearchCV
search = GridSearchCV(model1,params,cv=cv)
result=search.fit(x_sc,y)
result.best_params_
```

```python

model2 = Ridge(alpha=10)
model2.fit(xtrain,ytrain)
ypred2=model2.predict(xtest)
r2_score(ytest, ypred2)
```

```python
#applying lasso 
from sklearn.linear_model import Lasso
lr = Lasso()
lr.fit(xtrain,ytrain)
```

```python
predict2 = lr.predict(xtest)
print(r2_score(ytest,predict2))
```

```python
print(mean_absolute_error(ytest,predict2))
```
